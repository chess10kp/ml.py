```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = torch.arange(-5, 5, 0.1).view(-1, 1) # col vec of tensor

func = - 5 * X
Y = func + 0.4 * torch.randn(X.size())
```

```python
plt.plot(X.numpy(), Y.numpy(), "b+", label="Y")
plt.plot(X.numpy(), func.numpy(), "r", label="func")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
weights = torch.Tensor(-10, dtype=torch.float32, requires_grad=True)
biases = torch.Tensor(-10, dtype=torch.float32, requires_grad=True)

def forward(X) -> torch.Tensor:
    return weights * X + biases

def loss(prediction) -> torch.Tensor:
    return torch.mean((prediction - Y) ** 2) 

losses = []
for i in range(20):
    pred = forward(X)
    loss = loss(pred)
    loss.backward()

    weights.data = weights.data - 0.1 * weights.grad.data
    biases.data = biases.data - 0.1 * biases.grad.data
    weights.grad.data.zero_()
    weights.grad.data.zero_()
```
